```python
from loguru import logger
import pandas as pd 
```

```python
jcr = pd.read_csv('JCR2024-UTF8.txt', encoding='utf-8')
chin = pd.read_csv('FQBJCR2025-UTF8.csv', encoding='utf-8')
```

```python
jcr.head(3)
```

```python
chin.head(3)
```

```python
(
    jcr
    .merge(right=chin, left_on='Journal', right_on='Journal', how='left')
    .replace({'IF(2024)': {'<0.1': 0}})
    .astype({'IF(2024)': 'float'})
    .sort_values(by='IF(2024)', ascending=False)
    .query('`IF(2024)` >= 1.0')
    [['Journal', 'IF(2024)', 'Review', 'Open Access', '大类', '大类分区', '小类1', '小类1分区']]
    .to_excel('Summary_JCR2024-IF.xlsx', index=False)
)
logger.info(f'Exported Summary_JCR2024-IF.xlsx successfully!')
```
